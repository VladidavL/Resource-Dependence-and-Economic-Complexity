```python
import pandas as pd
import numpy as np
import world_bank_data as wb
```

```python
# accessing total exports data
wb.search_indicators('Exports of goods and services')
```

```python
j1 = wb.get_series("NE.EXP.GNFS.CD", id_or_value='id', simplify_index=True)
j2 = wb.get_series("NE.EXP.GNFS.KD", id_or_value='id', simplify_index=True)
```

```python
i1 = pd.DataFrame(j1)
i2 = pd.DataFrame(j2)
i1.to_csv("i1")
i2.to_csv("i2")

path1 = pd.read_csv("i1")
path2 = pd.read_csv("i2")

k1 = pd.DataFrame(path1)
k2 = pd.DataFrame(path2)
```

```python
k1["map"] = k1.index
k2["map"] = k2.index

inst = k1.values
for i in range(len(inst)):
    inst[i][3] = str(inst[i][0]) + str(inst[i][1])
    
l1 = pd.DataFrame(inst, columns= ("country_code", "year", "t_exp", "map"))

inst = k2.values
for i in range(len(inst)):
    inst[i][3] = str(inst[i][0]) + str(inst[i][1])
    
l2 = pd.DataFrame(inst, columns= ("country_code", "year", "t_exp_cons", "map"))
```

```python
p1 = l1[["t_exp", "map"]]
p2 = l2[["t_exp_cons", "map"]]

exports = pd.merge(p1, p2, how="outer", on="map")
exports.head(3)
```

```python
exports.to_csv("exports.csv")
```
