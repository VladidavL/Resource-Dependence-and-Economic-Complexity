```python
"""
Drawing ACF and PACF graphs for choosing lags

Result: first-order autocorrelation for each modofication of economic complexity index

"""
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
os.chdir("C:\\Users\\Vladimir.Grigoryev\\Documents\\Python Scripts\\eci")
```

```python
path = pd.read_csv("final_dataset.csv")
dt = pd.DataFrame(path)
dt.drop("Unnamed: 0", axis = 1, inplace=True)
dt.head()
```

```python
filt = dt.code == "RUS"
rus = dt[filt]
rus.head()
```

```python
filt = dt.code == "NOR"
nor = dt[filt]
nor.head()
```

```python
plt.style.use('seaborn-whitegrid')

for column in rus.columns:
    if rus[column].dtypes == "float64":
        x = rus.year
        y = rus[column]
        plt.figure()
        plt.xlabel("year")
        plt.ylabel(column)
        print(plt.plot(x, y, 'o', color='black'))
```

```python
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.graphics.tsaplots import plot_pacf
```

```python
series = pd.Series(data=nor["eci1"].dropna())
```

```python
plt.acorr(series, maxlags=10)
plt.xlabel('Lag')
plt.ylabel('Autocorrelation')
plt.show()
```

```python
series = pd.Series(data=nor["eci1"].dropna())
plt.figure()
plot_acf(series, lags=10)
plt.ylabel("ACF Score")
plt.xlabel("Lag")
plt.show()
plt.savefig('acf_nor.png')

plt.figure()
plot_pacf(series, lags=10)
plt.ylabel("PACF Score")
plt.xlabel("Lag")
plt.show()
plt.savefig('pacf_nor.png')
```
