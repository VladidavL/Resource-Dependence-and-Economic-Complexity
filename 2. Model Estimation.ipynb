```python
# code for running dynamic panel with FOD transformation
def model(endog, var_of_interest, bound, df):
    iter_res = []
    for num in bound:
        bv = df[df[var_of_interest] > num]
        # transformig data using forward orthogonal deviations
        dh = bv
        dh["T"] = 0
        dh["t"] = 0

        for i in dh.index:
            filt = dh.code == dh.at[i, "code"]
            inst = dh[filt]
            dh.at[i, "T"] = len(inst)
        
        dl = dh[[endog, var_of_interest, 'gdp_pc', 'population']]
        for col in dl.columns:
            for i in dh.index:
                if dh.at[i, "T"] == dh.at[i, "t"]:
                    dh.at[i, col] = np.nan
                elif dh.at[i, "T"] != dh.at[i, "t"]:
                    filt = dh.code  == dh.at[i, "code"]
                    inst = dh[filt]
                    c = 1/(dh.at[i, "T"] - dh.at[i, "t"])
                    b = dh[col][i+1:dh.at[i, "T"]].sum()
                    a = math.sqrt((dh.at[i, "T"] - dh.at[i, "t"])/(dh.at[i, "T"] - dh.at[i, "t"] + 1))
                    dh.at[i, col] = a*(dh.at[i, col] - c*b)
     
    #Creating predicted values for endogenous variable (using instruments)
        countries = list(dh.code.unique())
        predicted = []
        actual = []
        previous = []
        for name in countries:
            filt = dh["code"] == name
            series = pd.Series(dh[filt][endog].dropna())
            if len(series) > 1:
                X = series.values
                model = AR(X)
                model_fit = model.fit(maxlag=1)
                X_hat = model_fit.predict()
                actual.extend(X)
                predicted.extend([np.nan, np.nan])
                predicted.extend(X_hat)
                predicted.pop(-1)
                previous.extend([np.nan])
                previous.extend(X)
                previous.pop(-1)

        joint1 = pd.DataFrame([actual, predicted, previous]).T
        joint1.columns = [endog, str(endog+"_hat_t1"), str(endog+"_t1")]
        dh = pd.merge(dh, joint1, how="outer", on=endog)
    
    #Running Model
        data = dh
        year = pd.Categorical(data.year)
        data = data.set_index(["code", 'year'])
        data['year'] = year
    
        controls = ["population", "gdp_pc", str(endog+"_hat_t1")]
        exog_vars = [var_of_interest]
        exog = sm.add_constant(data[exog_vars  + controls])
        mod1 = PanelOLS(data[endog], exog)
        fe_res = mod1.fit()
        
        return fe_res
```
