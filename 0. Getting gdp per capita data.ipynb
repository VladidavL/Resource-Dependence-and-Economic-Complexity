```python
# adding oil (and other liquids) production
path = pd.read_csv("API_NY.GDP.PCAP.CD_DS2_en_csv_v2_10576699.csv", header = 2)
tr = pd.DataFrame(path)
tr.head()
```

```python
tr.drop(["Indicator Code", "Indicator Name", "Country Name", "Unnamed: 63"], axis=1, inplace=True)
```

```python
tr.tail()
```

```python
inst = tr.values
inst
```

```python
data = np.array(["code", "year", "gdp_pc"])
years = np.arange(1960,2019)
for i in range(len(inst)):
    name = inst[i][0]
    country = [name for x in range(len(years))]
    series = inst[i][1:]
    block = np.vstack((country, years, series))
    data = np.vstack((data, block.T))
data
```

```python
er = pd.DataFrame(data)
headers = er.iloc[0]
et  = pd.DataFrame(er.values[1:], columns=headers)
et.head()
```

```python
et["map"] = et.index
inst = et.values

for i in et.index:
    inst[i][3] = str(inst[i][0]) + str(inst[i][1])
    
ei = pd.DataFrame(inst, columns=("code", "year", "gdp_pc", "map"))
ei.head()
```

```python
eo = ei[["map", "gdp_pc"]]
```

```python
eo.to_csv("GDP_pc.csv")
```
