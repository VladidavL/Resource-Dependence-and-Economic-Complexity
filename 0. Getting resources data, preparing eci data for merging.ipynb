```python
# 2. Accessing eci datasets and creating variable "map" to help merge with wb data later
# preparing eci datasets
# preparing eci1 (detailed)
path1 = pd.read_csv("eci1.csv")
eci1 = pd.DataFrame(path1)
```

```python
eci1.head()
```

```python
# adding variable "map"
eci1["map"] = eci1.index
```

```python
# turning dataframe to matrix 
inst = eci1.values
```

```python
# turning matrix to dataframe back and assigning variable names again (as names got lost after tutned to matrix)
eci1 = pd.DataFrame(inst, columns = ["smth", "location_code", "year", "eci1", "map"])
```

```python
# dropping unuseful column
eci1.drop("smth", axis=1, inplace=True)
```

```python
eci1.head()
```

```python
inst = eci1.values
for i in range(len(inst)):
    inst[i][3] = str(inst[i][0]) + str(inst[i][1])
    
eci1 = pd.DataFrame(inst, columns = ("location_code", "year", "eci1", "map"))
eci1.drop(["location_code", "year"], axis = 1, inplace=True)
eci1.head()
```

```python
# preparing eci2 (quickly)
path = pd.read_csv("eci2.csv")
eci2 = pd.DataFrame(path)

eci2["map"] = eci2.index

inst = eci2.values

for i in range(len(inst)):
    inst[i][4] = str(inst[i][1]) + str(inst[i][2])
    
eci2 = pd.DataFrame(inst, columns = ("smth", "location_code", "year", "eci2", "map"))
eci2.drop(["smth", "location_code", "year"], axis = 1, inplace=True)
eci2.head()
```

```python
# preparing eci3 (quickly)
path = pd.read_csv("eci3.csv")
eci3 = pd.DataFrame(path)

eci3["map"] = eci3.index

inst = eci3.values

for i in eci3.index:
    inst[i][4] = str(inst[i][1]) + str(inst[i][2])
    
eci3 = pd.DataFrame(inst, columns = ("smth", "location_code", "year", "eci3", "map"))
eci3.drop(["smth", "location_code", "year"], axis = 1, inplace=True)
eci3.head()
```

```python
eci1.to_csv("eci1.csv")
eci2.to_csv("eci2.csv")
eci3.to_csv("eci3.csv")
```
