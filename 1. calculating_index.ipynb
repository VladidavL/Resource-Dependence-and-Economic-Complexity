```python
import pandas as pd
import numpy as np

path1 = pd.read_csv("data1.csv")
data1 = pd.DataFrame(path1)

path2 = pd.read_csv("data2.csv")
data2 = pd.DataFrame(path2)

path3 = pd.read_csv("data3.csv")
data3 = pd.DataFrame(path3)
```

```python
# working with data1
!pip install ecomplexity
from ecomplexity import ecomplexity

data = data1[['year','location_code','sitc_product_code','export_value']]
trade_cols = {'time':'year', 'loc':'location_code', 'prod':'sitc_product_code', 'val':'export_value'}
eci = ecomplexity(data, trade_cols)
data = eci
```

```python
eci.head(5)
```

```python
test = eci[["location_code", "year", "eci"]]
test.head()
```

```python
test = eci[["location_code", "year", "eci"]]

ylist = test.year.unique()

dt = test.drop(test.index)

for yar in ylist:
    filt = test.year == yar
    test_1 = test[filt]
    test_1 = test_1.drop_duplicates(subset = "location_code", keep = "first")
    dt = pd.concat([dt, test_1], axis = 0)
dt.head(5)
dt.sort_values(by = ["year"])

eci1 = dt
eci1.to_csv("eci1.csv")
```
